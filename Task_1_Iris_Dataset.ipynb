```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris.csv')
df.shape
df.columns
df.head()
plt.figure()
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.show()
plt.figure()
plt.hist(df['sepal_length'], bins=20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
plt.figure()
sns.boxplot(data=df, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.show()
print("""## Conclusion
In this analysis, we explored the Iris dataset using basic data inspection and visualization techniques.
Scatter plots showed clear separation between species based on petal measurements, while histograms and
box plots helped understand data distribution and variability. Petal features appear more useful than
sepal features for distinguishing flower species.""")
```
